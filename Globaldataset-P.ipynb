# USING HYPOTHESIS TESTING METHOD(ANNOVA ONEWAY)

```python
from scipy.stats import f_oneway
```

```python
i='Product ID'
v=[d_num_vals['TotalProfit'][d_cat_vals['Product ID']==category] for category in d_cat_vals['Product ID'].unique()]
f_stat,p_value = f_oneway(*v)
print(f"column : {i} , f-statical : {f_stat} , p_value : {p_value}")
```

```python
c_col=[]
nc_col=[]
for i in d_cat_vals.columns:
    if i == 'tp':
        continue
    else:
        v=[d_num_vals['TotalProfit'][d_cat_vals[i]==category] for category in d_cat_vals[i].unique()]
        f_stat,p_value = f_oneway(*v)
        print(f"column : {i} , f-statical : {f_stat} , p_value : {p_value}")
        if p_value < 0.05:
            c_col.append(i)
        else:
            nc_col.append(i)
```

# REMOVING OUTLIER (IQR)

```python
def remove_outlier_iqr(d,columns):
    for c in columns:
        Q1=d[c].quantile(0.25)
        median=d[c].quantile(0.50)
        Q3=d[c].quantile(0.75)
        IQR=Q3-Q1
        Lower=Q1-0.5*IQR
        Higer=Q3+0.5*IQR

        #remove outlier
        d_ro=d[(d[c] >= Lower) & (d[c] <= Higer)]
    return d_ro
```

```python
check_col=['Row ID',
 'Order Date',
 'Ship Date',
 'Sales',
 'Quantity',
 'Discount',
 'Profit',
 'Shipping Cost',
 'ActualCost',
 'TotalProfit']
```

```python
remove_outlier=remove_outlier_iqr(d_num_vals,check_col)
print("\n Orginal Dataframe:")
print(d_num_vals.shape)
print("\n After Remove outlier Dataframe")
print(remove_outlier.shape)
```

# Linear Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error,root_mean_squared_error
```

```python

#train and test the Dataset
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

#model building
lr=LinearRegression()
lr.fit(x_train,y_train)

#check score for train and test data
trs=lr.score(x_train,y_train)
tes=lr.score(x_test,y_test)

#predicted value of x test
y_pred=lr.predict(x_test)

#finding metric for regression
mse=mean_squared_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)

#print all above metric and score
print(f"Train score :{trs}")
print(f"Test score :{tes}")
print(f"R Score :{r2:.2f}")
print(f"MSE :{mse:.2f}")
print(f"MAE :{mae:.2f}")
```
